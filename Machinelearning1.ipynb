importing dataset

```python
import pandas as pd
ab = pd.read_csv('Iris.csv')
print(ab.head())
sepal_width = ab['SepalWidthCm']

# Mean
mean_value = sepal_width.mean()
print("Mean of SepalWidthCm:", mean_value)

# Median
median_value = sepal_width.median()
print("Median of SepalWidthCm:", median_value)

# Mode
mode_value = sepal_width.mode()
print("Mode of SepalWidthCm:", mode_value)

#Standard Deviation
std_deviation = sepal_width.std()
print("Standard Deviation of SepalWidthCm:", std_deviation)
```

Create a line chart

```python
import pandas as pd
import matplotlib.pyplot as plt

ab = pd.read_csv('Iris.csv')

plt.figure(figsize=(10, 6))
plt.plot(ab['Id'], ab['SepalWidthCm'], marker='o', linestyle='-', color='blue', label='SepalWidthCm')

plt.title('Sepal Width Across Samples')
plt.xlabel('Sample Id')
plt.ylabel('Sepal Width (cm)')
plt.grid(True)
plt.legend()
plt.tight_layout()

plt.show()
```
